## Imports

```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

1. Lance une requête sur la page web indiquée ci-dessus.

```python
# Test de la requête web
page = requests.get("https://www.chucknorrisfacts.fr/facts/top/1")
soup = BeautifulSoup(page.content, 'html.parser')
page
```

2. Quel est le response code ?

```python
print(page)
```

3. Trouve la balise permettant d'identifier le bloc contenant une blague. Combien y en a-t-il ? (il devrait y en avoir 20). Affiche le 8e bloc de blague grâce à l'attribut .text.

```python

blagues = soup.find_all(class_="card-text")
print(f"Nombre de blagues: {len(blagues)}")
```

4. Chacun des éléments de ton itérable est lui-même une mini-soupe. Utilise la méthode .find avec la balise appropriée pour isoler uniquement la 8e blague, sans les autres éléments (donc sans la note, etc...)

```python
# La 8eme blague est vide
blagues[7].text
```

```python
# Affichage de la 9eme blague
blagues[8].text
```

5. Fais de même pour isoler uniquement la note de la 8e blague

```python
# Isolation de la note en utilisant Regex
spans = soup.find_all('span', id=re.compile(r"moyenne_"))
print(re.search(r"[\d.]+", spans[7].text).group())
```

6. Crée un dictionnaire vide, puis crée une boucle qui viendra enregistrer dans ce dictionnaire chaque blague en clé et chaque note correspondante en valeur.

```python
mon_dictionnaire = dict()
for i in range(0, len(blagues)):
    mon_dictionnaire[soup.find_all(class_="card-text")[i].text] = re.search(r"[\d.]+", spans[i].text).group()
mon_dictionnaire
```

7. Transforme ce dictionnaire en DataFrame de 2 colonnes : blague et note. Il doit avoir 20 lignes : une par blague.

```python
df = pd.DataFrame(mon_dictionnaire.items(), columns=['Blagues', 'Notes'])
df
```

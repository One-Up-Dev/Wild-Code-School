```python
#Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = sns.load_dataset('iris')
iris.sample(10)
```

```python
X = iris.iloc[:,[0,1,2,3]]
```

```python
Kmeans_5 = KMeans(n_clusters=5)
y_kmeans5 = Kmeans_5.fit_predict(X)
```

```python
print(y_kmeans5)
```

```python
Kmeans_5.cluster_centers_
```

```python
Error =[]
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i).fit(X)
    kmeans.fit(X)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 11), Error)
plt.title('Elbow method')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

```python
Kmeans_3 = KMeans(n_clusters=3)
y_kmeans3 = Kmeans_3.fit_predict(X)+1
```

```python
y_kmeans3
```

```python
Kmeans_3.cluster_centers_
```

```python
plt.scatter(X.petal_length, X.petal_width, c=y_kmeans3)
```

```python
iris['resultat'] = Kmeans_3.predict(X)+1
```

# **KMEANS - Clustering**

## Imports

```python
#Import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

---
## KMeans sur 3 Clusters

```python
iris = sns.load_dataset('iris')
iris.head(4)
```

```python
X = iris.iloc[:,[0,1,2,3]]
X.sample(4)
```

```python
# Choix du modèle
model = KMeans(n_clusters=3, random_state=42)
# entrainement + prediction sur le dataset X
resultat = model.fit_predict(X)
# Affichage du résultat
resultat
```

```python
iris['resultat'] = model.predict(X)+1
X['resultat'] = model.predict(X)+1
```

```python
X
```

```python
iris.sample(4)
```

```python
model.cluster_centers_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,3))
ax1.scatter(X.petal_length, X.petal_width, c=X["resultat"], s=10)
ax1.scatter(model.cluster_centers_[:, 2], model.cluster_centers_[:, 3], c="red", marker ="X", s=100)
ax1.set_title("3 clusters")
ax2.scatter(iris.petal_length, iris.petal_width, c=iris["species"].factorize()[0], s= 10)
ax2.set_title("3 original species")
plt.show()
```

### Méthode Silhouette

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
K_range = range(2, 10) # La silhouette commence à 2 clusters minimum

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X)
    score = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(score)

plt.plot(K_range, silhouette_scores, marker='o')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Score de Silhouette')
plt.title('Méthode de la Silhouette')
plt.show()

# Le nombre optimal de cluster est la valeur la plus elevé du score
# Ici c'est 3 (on ne prend pas en compte la 1ère valeur)
```
